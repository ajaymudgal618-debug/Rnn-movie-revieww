```python
# sentencs

sent= [ "the glass of milk",
    "the glass of juice",
    "the  cup of tea",
    "i am good boy",
    "i am good developer",
    "understand the maening of words",
    "your videos are good"]
```

```python
sent
```

```python
from tensorflow.keras.preprocessing.text import one_hot

voc_size = 1000
onehot_repr = [one_hot(words,voc_size) for words in sent]
onehot_repr
```

```python
## word embedding
from  tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_len= 6
embedded_docs= pad_sequences(onehot_repr,padding='pre',maxlen=sent_len)
embedded_docs
```

```python
## feature represntation
dim=10
model= Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_len))
model.compile('adam','mse')
model.summary()
```

```python
predict=model.predict(embedded_docs[0])
predict
```
